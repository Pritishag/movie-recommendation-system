```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies.info()
```

```python
# Required columns
# --> genres
# --> id
# --> keywords
# --> title
# --> overview
# --> cast (actors)
# --> crew (director name)

movies = movies[['movie_id', 'original_title', 'genres', 'keywords', 'overview', 'cast', 'crew']]
```

```python
movies.head(1)
```

```python
movies.rename(columns={"original_title": "title"}, inplace = True)
```

```python
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies['genres'][0]
```

```python
# we just need the keywords of the genres
import ast

def modify(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(modify)
```

```python
movies.head(1)
```

```python
movies['keywords'][0]
```

```python
movies['keywords'] = movies['keywords'].apply(modify)
```

```python
movies.head(1)
```

```python
movies['cast'][0]
```

```python
def cast_modify(obj):
    l = []
    count = 0
    for i in ast.literal_eval(obj):
        l.append(i['name'])
        count += 1
        if count == 3:
            break
    return l
```

```python
movies['cast'].apply(cast_modify)
```

```python
movies['cast'] = movies['cast'].apply(cast_modify)
```

```python
movies['crew'][0]
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            print(i['name'])
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(1)
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies['overview']
```

```python
movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['keywords'] + movies['genres'] + movies['cast'] + movies['crew']
```

```python
movies['tags'][0]
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df.head(1)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags']
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
sorted(enumerate(similarity[0]), reverse=True, key=lambda x:x[1])[1:6]
    
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(enumerate(distances), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]]['title'])
```

```python
recommend('Avatar')
```
